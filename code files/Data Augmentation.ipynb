Define a function that performs flip, rotation (of 90, 180 and, 270 degrees), and mirror operation to augment the dataset.

```
image_dir = "damaged_structures_detector/images"  # path to the folder that contains the images of the dataset

def augment_images(directory=None, rotate_angle = 90 ):  # the function takes 2 inputs. 1) the folder that contains the images you want to augment, with None as the default input. 2) an angle of rotation with 90 degrees being the default input (the angle will be multiplied by 2 until a full rotation is performed, i.e. 90 means that rotations of 90, 180 and,270 degrees will be performed).
    filenames = os.listdir(directory)

    for filename in filenames:
      if '_' not in filename:
        angle = rotate_angle
        while angle < 360:
            im = Image.open(os.path.join(directory,filename))
            im = im.rotate(angle)
            new_filename = os.path.splitext(filename)[0]+'_rotate_%s.jpg'%angle  # create new file that contains the angle of rotation on its title
            im.save(os.path.join(directory,new_filename))
            angle += rotate_angle

        im = Image.open(os.path.join(directory,filename))
        im = ImageOps.flip(im)
        new_filename = os.path.splitext(filename)[0]+'_flip.jpg' # create new file that contains the word flip on its title
        im.save(os.path.join(directory,new_filename))
        print(os.path.join(directory,new_filename))
        im = Image.open(os.path.join(directory,filename))
        im = ImageOps.mirror(im)
        new_filename = os.path.splitext(filename)[0]+'_mirror.jpg' # create new file that contains the word mirror on its title
        im.save(os.path.join(directory,new_filename))


```

Use the function of the previous cell to perform data augmentation.

```
augment_images(os.path.join(image_dir,'val','not_destroyed'))
```
